## Imports

```python
import customized_chart
print(customized_chart.__version__)
from customized_chart import *
from random import Random
%matplotlib inline
```

## Test data

```python
movies = {
    "The Shawshank Redemption": 2600000,
    "The Dark Knight": 2600000,
    "Batman Begins": 1500000,
    "The Light Between Oceans": 57000,
    "The Empire Strikes Back": 1300000,
}

rnd = Random(42)
letters = {}
for i in range(65,91):
    letters[chr(i)] = rnd.randrange(8) + 1

rnd = Random(42)
costs = []
for i in range(0,33):
    costs.append(rnd.randrange(20) + 100 + i)
    
rnd = Random(42)
rental = {
    "values": [
        [(rnd.randrange(22) + 100 + i) for i in range(0,4)],
        [(rnd.randrange(15) + 40 + i/2) for i in range(0,4)],
        [(rnd.randrange(8) + 60 + i/3) for i in range(0,4)]
    ],
    "labels": ["2021-Q1", "2021-Q2", "2021-Q3", "2021-Q4"],
    "series": ["Bruce", "Robin", "Albert"]
}
```

## Example charts

### Pie

```python
pie_chart(movies, {
    "explode": ["The Dark Knight", "The Shawshank Redemption"], 
    "title": "IMDB number of ratings", 
    "shadow": True,
    "percent_format": "pct-2",
    "value_format": "prefix",
})
```

### Pie, specified total

```python
pie_chart(movies, {
    "explode": ["The Dark Knight", "The Shawshank Redemption"], 
    "title": "IMDB number of ratings", 
    "shadow": True,
    "percent_format": "pct-2",
    "value_format": "prefix-2",
    "total": sum([x for x in movies.values()])*2,
    "show_total": True,
})
```

### Pie, sorted data

```python
sorted_movies = sort(movies, field="value", reverse=True)
pie_chart(sorted_movies, {
    "title": "IMDB number of ratings", 
    "percent_format": "pct-1",
    "value_format": "prefix",
})
```

### Pie, other color map

```python
pie_chart(letters, opts={
    "show_counts": False, 
    "cmap": "viridis", 
    "percent_format": "pct-0",
    "label_color": "#000",
    "label_distance": 0.7,
})
```

### Pie, border between slices

```python
pie_chart(letters, opts={
    "show_counts": False, 
    "cmap": "viridis", 
    "percent_format": "pct-0",
    "label_color": "#000",
    "label_distance": 0.7,
    "border": ["white", 1],
})
```

### Pie, specified colors

```python
pie_chart(movies, {
    "title": "IMDB number of ratings", 
    "shadow": False,
    "percent_format": "pct-2",
    "value_format": "prefix",
    "cmap": ["#444444", "#666666", "#888888", "#AAAAAA", "#CCCCCC"],
})
```

### Pie, set max slices to show

```python
pie_chart(letters, opts={
    "show_counts": False, 
    "percent_format": "pct-0",
    "max_slices": 10,
})
```

### Bar

```python
bar_chart(movies, {
    "title": "IMDB number of ratings", 
    "value_format": "prefix",
    "label_rotation": 45,
    "horizontal": False,
    "grid": True,
    "bar_label": True,
    "y_label": "No ratings",
    "x_label": "Movie"
})
```

### Grouped bar

```python
grouped_bar_chart(rental, {
    "title": "Movie rental costs",
    "bar_labels": True,
    "value_format": "dec-2",
    "y_label": "USD",
    "x_label": "Quarter",
    "grid": True,
    "y_lim": (0,140),
})
```

### Line

```python
line_chart(costs, {
    "title": "My movie rental costs", 
    "other_label": "vriga",
    "label_rotation": 45,
    "grid": True,
    "y_label": "USD",
    "labels": "month 2020",
    "y_lim": (0,200),
    "trend": True,
})
```

### Multiple lines

```python
multi_line_chart(rental, {
    "cmap": "Paired",
    "trend": True,
    "title": "Movie rental costs",
    "y_lim": [0,140],
})
```
